```python
# make functions to apply on the dataframe
def continent(x):
    if x['Geography'] in ['France', 'Germany', 'Spain']:
        return 'Europe'
    else:
        return 'Correct me'
    
df_copy['Continent'] = df_copy.apply(continent, axis=1)
df_copy.head()
```

```python
# concat dataframes
df_1 = pd.DataFrame({
    "A": ["A1", "A2", "A3"],
    "B": ["B1", "B2", "B3"],
    "C": ["C1", "C2", "C3"],
})

df_2 = pd.DataFrame({
    "A": ["A4", "A5", "A6"],
    "B": ["B4", "B5", "B6"],
    "C": ["C1", "C4", "C3"],
})

df_concat = pd.concat([df_1, df_2])
df_concat
```

```python
# merge dataframes (inner join)
df_merge = pd.merge(df_1, df_2, on='C', how='inner', suffixes=('_left', '_right'))
df_merge
```

```python
# merge dataframes (left join)
df_merge = pd.merge(df_1, df_2, on='C', how='left', suffixes=('_left', '_right'))
df_merge
```

```python
# merge dataframes (right join)
df_merge = pd.merge(df_1, df_2, on='C', how='right', suffixes=('_left', '_right'))
df_merge
```

```python
# merge dataframes (outer join)
df_merge = pd.merge(df_1, df_2, on='C', how='outer', suffixes=('_left', '_right'))
df_merge
```

```python
# group by a column and apply an aggregation function
df_group = df_copy.groupby('Geography').agg({'Age': 'mean', 'Balance': 'sum'})
df_group
```
